```python
players = pd.read_csv(r"C:\Users\kannu\OneDrive\Documents\github repos\University\year-2\sem-4\ML\extras\project-2\Proto\players.csv")
del players["Rk"]
del players["Unnamed: 0"]
players["Player"] = players["Player"].str.replace("*", "", regex = False)
players.head(5)


```

Combining the players and teams dataframes.

```python
teams = pd.read_csv(r"C:\Users\kannu\OneDrive\Documents\github repos\University\year-2\sem-4\ML\extras\project-2\Proto\teams.csv")

# Here, this iterable command Checks for every row containing Division and deletes it
teams = teams[~teams["W"].str.contains("Division")].copy()
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
del teams["Unnamed: 0"]
teams
```

```python
sorted(teams["Team"].unique())
sorted(players["Tm"].unique())

nicknames = {}
with open(r"C:\Users\kannu\OneDrive\Documents\github repos\University\year-2\sem-4\ML\extras\project-2\Proto\nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name

nicknames

players["Team"] = players["Tm"].map(nicknames)

players
```
