```python
from collections import defaultdict


def intToOrdinal(n):
    ordDict = defaultdict(lambda: "th")
    ordDict.update({1: "st", 2: "nd", 3: "rd"})
    q, mod = divmod(n, 10)
    suffix = "th" if q % 10 == 1 else ordDict[mod]
    return f"{n}{suffix}"
```

```python
def subPlotVertical(img):
    n = int(input("Enter the number of splits: "))
    imgW = img.shape[0]

    fig, ax = plt.subplots(1, n, figsize = (10, 4), sharey = True, sharex = True)
    for i in range (n):
        ax[int(i)].imshow(img[:, int(i*imgW/n):int((i+1)*imgW/n)])
        ax[int(i)].set_title(intToOrdinal(i+1) + ' Split')
```

```python
subPlotVertical(lena)
```

```python
def subPlotHorizontal(img):
    n = int(input("Enter the number of splits: "))
    imgW = img.shape[0]

    fig, ax = plt.subplots(n, 1, figsize=(10, 4), sharex=True)
    for i in range(n):
        ax[int(i)].imshow(img[int(i*imgW/n):int((i+1)*imgW/n), :])
        ax[int(i)].set_title(intToOrdinal(i+1) + ' Split')
```

```python
subPlotHorizontal(lena)
```

```python
def subPlotRGB(img):
    fig, ax = plt.subplots(1, 3, figsize = (12, 4), sharey = True)
    
    ax[0].imshow(img[:, :, 0], cmap = 'Reds')
    ax[0].set_title('Red')
    
    ax[1].imshow(img[:, :, 1], cmap='Greens')
    ax[1].set_title('Green')
    
    ax[2].imshow(img[:, :, 2], cmap='Blues')
    ax[2].set_title('Blue')
```

```python
subPlotRGB(lena);
```

```python
lenaHSV = rgb2hsv(lena)


def subPlotHSV(img):
    fig, ax = plt.subplots(1, 3, figsize=(12, 4), sharey=True)

    ax[0].imshow(img[:, :, 0], cmap='hsv')
    ax[0].set_title('Hue')

    ax[1].imshow(img[:, :, 1], cmap='hsv')
    ax[1].set_title('Saturation')

    ax[2].imshow(img[:, :, 2], cmap='hsv')
    ax[2].set_title('Value')
    
subPlotHSV(lenaHSV)
```

```python
def subPlotGray(img):
    fig, ax = plt.subplots(1, 5, figsize=(12, 4), sharey=True)

    ax[0].imshow(img, cmap='gray')
    ax[0].set_title('Original')

    ax[1].imshow(img_as_uint(img > .25), cmap='gray')
    ax[1].set_title('Greater than 0.25')

    ax[2].imshow(img_as_uint(img > .50), cmap='gray')
    ax[2].set_title('Greater than 0.50')
    
    ax[3].imshow(img_as_uint(img > .75), cmap='gray')
    ax[3].set_title('Greater than 0.75')
    
    ax[4].imshow(img_as_uint(img > np.mean(lenaGray)), cmap='gray')
    ax[4].set_title('Greater than the Mean')

```

```python
lenaGray = rgb2gray(lena)
subPlotGray(lenaGray)
```
