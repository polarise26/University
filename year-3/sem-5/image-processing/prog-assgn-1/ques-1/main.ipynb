Importing Dependencies

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
img512 = Image.open("lena512.bmp")
```

```python
img512.size
```

```python
img512.format
```

```python
img256 = img512.resize((256, 256))
img256.save('lena256.bmp')
```

### Scaling up the images using the 6 different methods in PIL

```python
img512Nearest = img256.resize((512, 512), Image.NEAREST)
img512Bilinear = img256.resize((512, 512), Image.BILINEAR)
img512Bicubic = img256.resize((512, 512), Image.BICUBIC)
img512Lanczos = img256.resize((512, 512), Image.LANCZOS)
img512Box = img256.resize((512, 512), Image.BOX)
img512Hamming = img256.resize((512, 512), Image.HAMMING)
```

### Function to calculate PSNR

```python
def PSNR(og, compr, maxVal = 255):
    MSE = np.mean((np.array(og) - np.array(compr) ** 2))
    if MSE == 0:
        return 100
    PSNR = 10 * np.log10(maxVal ** 2 / MSE)
    return PSNR
```

```python
PSNR(img512, img512Box)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].imshow(img512Nearest, cmap = 'gray')
ax[0].set_ylabel('Nearest Neighbour')
ax[0].set_xlabel(PSNR(img512, img512Nearest))

ax[1].imshow(img512Bilinear, cmap='gray')
ax[1].set_ylabel('Bilinear')
ax[1].set_xlabel(PSNR(img512, img512Bilinear))

ax[2].imshow(img512Bicubic, cmap = 'gray')
ax[2].set_ylabel('Bicubic')
ax[2].set_xlabel(PSNR(img512, img512Bicubic))
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].imshow(img512Lanczos, cmap = 'gray')
ax[0].set_ylabel('Lanczos')
ax[0].set_xlabel(PSNR(img512, img512Lanczos))

ax[1].imshow(img512Box, cmap = 'gray')
ax[1].set_ylabel('Box')
ax[1].set_xlabel(PSNR(img512, img512Box))

ax[2].imshow(img512Hamming, cmap = 'gray')
ax[2].set_ylabel('Hamming')
ax[2].set_xlabel(PSNR(img512, img512Hamming));
```
