### Importing dependencies

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

from skimage import img_as_uint
from skimage.io import imshow, imread
from skimage.color import rgb2hsv, rgb2gray
```

```python
kidRed = Image.open("kidred.jpg")
redEye = Image.open("redeye.jpg")
redEyeBaby = Image.open("redeyebaby.jpg")
redEyeEffect = Image.open("redeyeeffect.png")

```

### Definition of function to remove red eyes:
I came up with the function values all by myself and settled on this after a bit of trial and error. 

```python
def redEyeRemover(img):
    w, h = img.size
    newImg = Image.new('RGB', (w, h))
    origMap = img.load()
    newMap = newImg.load()
    for x in range(w):
        for y in range(h):
            origPix = origMap[x, y]
            R = origPix[0]
            B = origPix[1]
            G = origPix[2]
            if R > (1.5 * max(G, B)):
                R = int((G + B)/1.5)
            newPix = (R, G, B)
            newMap[x, y] = newPix
    return newImg
```

```python
correctedImg = redEyeRemover(kidRed)
correctedImg.save("kidredREMOVED.jpg")

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(kidRed)
axs[0].set_title('Original Image')

axs[1].imshow(correctedImg)
axs[1].set_title('Corrected Image');
```

```python
correctedImg = redEyeRemover(redEye)
correctedImg.save("redeyeREMOVED.jpg")

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(redEye)
axs[0].set_title('Original Image')

axs[1].imshow(correctedImg)
axs[1].set_title('Corrected Image');
```
