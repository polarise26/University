```python
from sklearn.ensemble import IsolationForest
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt

# Load the dataset
data = load_wine()
X = data.data

# Initialize the Isolation Forest model
model = IsolationForest(n_estimators=100, contamination='auto', random_state=42)

# Fit the model to the data
model.fit(X)

# Predict the outlier scores for each data point
outlier_scores = model.score_samples(X)

# Plot the outlier scores against the data points
plt.scatter(range(len(X)), X[:, 0], c=outlier_scores, cmap='inferno')
plt.xlabel("Index")
plt.ylabel(f"{data.feature_names[0]}".title())
plt.colorbar()
plt.show()
```
